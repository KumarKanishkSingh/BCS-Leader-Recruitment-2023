```python
import pandas as pd
import random
import string
```

```python
# load the CSV file into a pandas dataframe
df = pd.read_csv('/content/train - train.csv')
```

```python
# create a list of negative sentiments
neg_sentiments = ["terrible", "bad", "frustrating", "disappointing", "difficult"]
```

```python
# iterate over the dataframe and randomly generate negative samples
neg_samples = []
for index, row in df.iterrows():
    # check if the label is 1 (positive)
    if row['label'] == 1:
        # generate a negative sample with opposite sentiment
        neg_text = random.choice(neg_sentiments) + " " + row['reason']
        neg_samples.append({'text': neg_text, 'label': 0})
```

```python
# convert the negative samples to a pandas dataframe and save to CSV
neg_df = pd.DataFrame(neg_samples)
neg_df.to_csv('neg_samples.csv', index=False)
neg_df
```

```python
column_name = df['reason']
common_column=df['text']
```

```python
merged_df = pd.merge(neg_df, column_name, on=common_column)
merged_df
```

```python
merged_df = merged_df.drop('reason', axis=1)
```

```python
merged_df = merged_df[['key_0', 'text', 'label']]
```

```python
merged_df.rename(columns = {'key_0':'text', 'text':'reason'}, inplace = True)
```

```python
merged_df
```

```python
final_df = pd.concat([df, merged_df], ignore_index=True)
```

```python
final_df.to_csv('Train.csv', index=False)
```
